Preparando a consulta ao Science Direct/Scopus. Como a API limita o tamanho da consulta, serão feitas diversas consulta cada uma retornando 200 resultados. A consulta "SRCTITLE(journal_name)" retorna os artigos publicados em um determinao journal ou conferência. Para saber o código de outros campos disponível na base Scopus é só acessar essa página: https://service.elsevier.com/app/answers/detail/a_id/11236/supporthub/scopus/

O objetivo dessa etapa é recuperar o SCOPUS ID dos artigos publicados em um determino journal.

```python
# Dado um termo para busca do journal ("query") e número máximo de resultados ("finish") 
# a função retorna uma lista de DOI
def get_ScopusIDs (query, finish):
    ScopusIDs = []       # Lista para armazenar os DOI

    start = 1
    step =  20     # Número de DOI retornado por consulta

    print('Realizando Query por journal: ', query)

    while start < finish:
        
        url = ("https://api.elsevier.com/content/search/scopus?start=" + 
               str(start) + 
               "&count=" + 
               str(step) + 
               "&query=" + 
               query + 
               "&apiKey=" + 
               api_key)

        # Requisição à API da Elesevier
        f = requests.get(url, proxies = proxies).text

        #Parsing nos resultados para identificar os ScopusIDs
        f = json.loads(f)
        try:
            n_entries = len(f["search-results"]["entry"])
            for entry in f["search-results"]["entry"]:
                try:
                    ScopusIDs.append(entry['dc:identifier'])
                except:
                    pass
        except:
            pass
        start = start + step

    return(ScopusIDs)
```

Após identificar a lista com todos os ScopusIDs que temos interesse é necessário utilizar a API da Elsevier para acessar cada um dos artigos e recuperar as informações importantes de cada um deles. Neste script nós recuperamos as seguintes informações: "DOI", "SCOPUS_ID", "Title", "Date", "Keywords", "Creator", "Abstract", "Text".

```python
#Dado um documento em XML com as informações do artigo, retornar uma Série com 
# as informações de DOI, Título, Data, Resumo, Palavras-chave e Autores.
def get_article(article):
    
    article = BeautifulSoup(article, "lxml")

    try:
        DOI = article.find("dc:identifier").get_text()[4:]
    except:
        DOI = ""
        
    try:
        SCOPUS_ID = article.find("scopus-id").get_text()
    except:
        SCOPUS_ID = ""
        
    try:
        Title = article.find("dc:title").get_text()
    except:
        Title = ""
    
    try:
        Date = article.find("prism:coverdate").get_text()    
    except:
        Date = ""
    
    Keywords = []
    Creators = []
    try:
        coredata = article.find("coredata")
        for data in coredata.descendants:
            try:
                if data.name == "dcterms:subject":
                    Keywords.append(data.get_text())
            except:
                pass
            try:
                if data.name == "dc:creator":
                    Creators.append(data.get_text())
            except:
                pass
    except:
        pass
    
    try:
        Abstract = article.find("dc:description").get_text()
    except:
        Abstract = ""
        
    try:
        Text = article.find("ce:sections").get_text()
    except:
        Text = "" 
        
            
    row = pd.Series([DOI, SCOPUS_ID, Title, Date, Keywords, Creators, Abstract, Text],
                    index=["DOI", "SCOPUS_ID", "Title", "Date", "Keywords", "Creator", "Abstract", "Text"])
            
    return(row)
```

```python
# Essa função uni as duas anteriores. Dado o nome de uma publicação, o número máximo de respostas e 
# o nome do arquivo de saída, a função grava na pasta "target_path" um arquivo JSON com as informações dos
# artigos recuperados.

def get_abstracts(query, query_count, name_file):
    #Criando dataframe para armazenar as informações recuperadas
    articles = { "DOI":[], "Title": [], "Abstract": [], "Text": []}
    articles = pd.DataFrame(data=articles)
    
    # Recupera os ScopusIDs
    ScopusIDs = get_ScopusIDs (query, query_count)

    print('Buscando os arquivos no site da Elsevier...')
    print('Total de artigos: ', len(ScopusIDs))
    n = 0
    #Para cada ScopusID identificado recuperar as informações do artigo
    for ScopusID in ScopusIDs:
        # Fazer uma requisição na API da Elsevier
        f = requests.get("https://api.elsevier.com/content/article/scopus_id/" +
                         ScopusID[9:] +
                         "?APIKey=" +
                         api_key, proxies=proxies)
        f = f.text
        # Para cada artigo recuperar as informações
        articles = articles.append(get_article(f), ignore_index=True)
        
        n = n + 1
        if n % 20 == 0:
            print ('Artigos: ', n)

            #Gravar o arquivo em formato JSON        
            arquivo_saida = os.path.join(target_path, name_file+ ".json")
            print('Escrevendo arquivo: ', arquivo_saida)
            articles.to_json(arquivo_saida)
    return(articles)           
```
